```python
##Neuron
## Input -> Weight -> Summation -> Activation Function(Binary 0 or 1 based on threshold) -> Output
## Weight and Threshold can be modified by user
```

```python
x_input=[0.1,0.5,0.2]
w_weight=[0.4,0.3,0.6]
threshold=0.2       #For 0.5 output=0   #For above inp and weight the weighted sum value is 0.31  #ie. if thresh<w_sum=1 else when thresh>w_sum=0
```

```python
def perceptron():
  weighted_sum=0
  for x,w in zip(x_input,w_weight):
    weighted_sum+=x*w
    print(weighted_sum)
  return step(weighted_sum)
```

```python
## Step function(code) is taken as activation function(concept)
def step(weighted_sum):
  if weighted_sum>threshold:
    return 1
  else:
    return 0
```

```python
output=perceptron()
print("Output = ", output)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
```

```python
x=pd.read_csv('titanic_df1.csv')
y=pd.read_csv('titanic_target.csv')
x
```

```python
y.drop(columns=y.columns[0], axis=1,  inplace=True)
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
import numpy as np
```

```python
##Prediction (y`) = 1 if Wx+b > 0 and 0 if Wx+b ≤ 0
## w1*x1+w2*x2+b, initializing w1, w2, as 1 and b as –1

def step(w_sum):
  if w_sum>0:
    return 1
  else :
    return 0
```

```python
# AND Logic Function
# w1 = 1, w2 = 1, b = -1
def ANDperc(x):
  weight=np.array([1,1])
  b=-1
  return perceptron1(x,weight,b)
```

```python
def perceptron1(x,w,b):
  w_sum1=np.dot(x,w)+b
  res=step(w_sum1)
  return res
```

```python
# design Perceptron Model
def perceptron2(x, w, b):
  w_sum1=np.dot(x,w)+b
  res=step(w_sum1)
  return res

# OR Logic Function
# w1 = 2, w2 = 2, b = -1
def OR(x):
	w = np.array([2, 2])
	b = -1
	return perceptron2(x, w, b)

# testing the Perceptron Model
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([1, 1])

print("OR({}, {}) = {}".format(0, 0, OR(test1)))
print("OR({}, {}) = {}".format(0, 1, OR(test2)))
print("OR({}, {}) = {}".format(1, 0, OR(test3)))
print("OR({}, {}) = {}".format(1, 1, OR(test4)))
```

```python
# design Perceptron Model
def perceptron3(x, w, b):
  w_sum1=np.dot(x,w)+b
  res=step(w_sum1)
  return res

# NOT Logic Function
# w1 = -1, b = 1
def NOT(x):
	w = np.array([-1])
	b = 1
	return perceptron3(x, w, b)

# testing the Perceptron Model
test1 = np.array([0])
test2 = np.array([1])


print("NOT({}) = {}".format(0, NOT(test1)))
print("NOT({}) = {}".format(1, NOT(test2)))
```

```python
# design Perceptron Model
def perceptron4(x, w, b):
  w_sum1=np.dot(x,w)+b
  res=step(w_sum1)
  return res

# NOR Logic Function
# w1 = -1, w2 = -1, b = 1
def NOR(x):
	w = np.array([-1, -1])
	b = 1
	return perceptron4(x, w, b)

# testing the Perceptron Model
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([1, 1])

print("NOR({}, {}) = {}".format(0, 0, NOR(test1)))
print("NOR({}, {}) = {}".format(0, 1, NOR(test2)))
print("NOR({}, {}) = {}".format(1, 0, NOR(test3)))
print("NOR({}, {}) = {}".format(1, 1, NOR(test4)))
```

```python
# design Perceptron Model
def perceptron5(x, w, b):
  w_sum1=np.dot(x,w)+b
  res=step(w_sum1)
  return res

# NAND Logic Function
# w1 = -1, w2 = -1, b = 2
def NAND(x):
	w = np.array([-1, -1])
	b = 2
	return perceptron5(x, w, b)

# testing the Perceptron Model
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([1, 1])

print("NAND({}, {}) = {}".format(0, 0, NAND(test1)))
print("NAND({}, {}) = {}".format(0, 1, NAND(test2)))
print("NAND({}, {}) = {}".format(1, 0, NAND(test3)))
print("NAND({}, {}) = {}".format(1, 1, NAND(test4)))
```

```python
def XOR(x):
    y1 = ANDperc(x)
    y2 = OR(x)
    y3 = NOT(y1)
    final_x = np.array([y2, y3])
    finalOutput = ANDperc(final_x)
    return finalOutput

# testing the Perceptron Model
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([1, 1])

print("XOR({}, {}) = {}".format(0, 0, XOR(test1)))
print("XOR({}, {}) = {}".format(0, 1, XOR(test2)))
print("XOR({}, {}) = {}".format(1, 0, XOR(test3)))
print("XOR({}, {}) = {}".format(1, 1, XOR(test4)))
```
