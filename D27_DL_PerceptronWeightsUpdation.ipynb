```python
##Neuron
## Input -> Weight -> Summation -> Activation Function(Binary 0 or 1 based on threshold) -> Output
## Weight and Threshold can be modified by user
```

```python
x_input=[0.1,0.5,0.2]
w_weight=[0.4,0.3,0.6]
threshold=0.2       #For 0.5 output=0   #For above inp and weight the weighted sum value is 0.31  #ie. if thresh<w_sum=1 else when thresh>w_sum=0
```

```python
def perceptron():
  weighted_sum=0
  for x,w in zip(x_input,w_weight):
    weighted_sum+=x*w
    print(weighted_sum)
  return step(weighted_sum)
```

```python
## Step function(code) is taken as activation function(concept)
def step(weighted_sum):
  if weighted_sum>threshold:
    return 1
  else:
    return 0
```

```python
output=perceptron()
print("Output = ", output)
```

# Titanic Dataset


```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
```

```python
x=pd.read_csv('titanic_df1.csv')
y0 = pd.read_csv('titanic_target.csv', squeeze=True)
x.drop(columns=x.columns[0], axis=1,  inplace=True)
y0
```

```python
selected_columns = ['Survived']
y = y0[selected_columns].copy()
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(len(X_train),len(X_test),len(y_train),len(y_test))
```

```python
import numpy as np
import pandas as pd


class Perceptron:
    def __init__(self,lr_rate, epochs):
        self.wt = None
        self.b = None
        self.lr_rate = lr_rate
        self.epochs = epochs

    def activation(self,z):
      return np.heaviside(z,0)

    def fit(self,X_train,y_train):
      n_samples,n_features=X_train.shape
      self.wt=np.zeros(n_features)
      self.b=0

      for epoch in range(self.epochs):
        for i in range(n_samples):
          z=np.dot(X_train.iloc[i],self.wt)+self.b
          y_pred=self.activation(z)
          #Update weights and bias
          self.wt+=self.lr_rate*(y_train.iloc[i]-y_pred)*X_train.iloc[i]
          self.b+=self.lr_rate*(y_train.iloc[i]-y_pred)

    def predict(self,X_train):
      z=np.dot(X_train.values,self.wt)+self.b
      return self.activation(z)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


perceptron=Perceptron(0.01,10)
perceptron.fit(X_train,y_train)
pred=perceptron.predict(X_test)

accuracy = accuracy_score(y_test,pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, pred))
```
